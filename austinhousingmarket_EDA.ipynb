#### Filter the dataset to the needs of our client

```python
import pandas as pd

df_subset= df[
    (df['propertyTaxRate'] == 1.98) & 
    (df['hasView'] == 1) & 
    (df['numOfBedrooms'] >= 2) &  
    (df['latestPrice'] <= 1000000) &
    (df['avgSchoolRating'] >= 4) &
    (df['homeType'] != 'Vacant Land')]

print(df_subset.head())
print(f"Number of observations meeting all specified conditions: {len(df_subset)}")
```

#### Initial exploration with latestPrice to see if any of the variables in this first section of EDA relates to price in any way.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

variables = ['latitude', 'longitude', 'lotSizeSqFt', 'livingAreaSqFt']

# initialize list for storage of summaries
regression_summaries = []

# perform regression and print summary
def perform_regression(variable):
    X = df_subset[[variable]]
    y = df_subset['latestPrice']
    
    model = LinearRegression().fit(X, y)
    intercept = model.intercept_
    coefficients = model.coef_
    r_squared = model.score(X, y)
    
    # add summaries to initialized list
    regression_summaries.append({
        'Variable': variable,
        'Intercept': intercept,
        'Coefficient': coefficients[0],
        'R-squared': r_squared})
    
    print(f"Variable: {variable}")
    print(f"Intercept: {intercept}")
    print(f"Coefficient: {coefficients[0]}")
    print(f"R-squared: {r_squared}\n")

for variable in variables:
    perform_regression(variable)

# convert summaries to dataframe
summary_df = pd.DataFrame(regression_summaries)
print(summary_df)
```

#### This shows us that among the numerical variables that were selected, livingAreaSqFt has a relatively high R-squared score, which tells us that 29.5% of the variance in latestPrice is able to be explained by an increase or decrease in livingAreaSqFt.

#### Let's make a new column, named pricePerSqFt that divides the latestPrice value by the livingAreaSqFt value.

```python
df_subset['pricePerSqFt'] = df_subset['latestPrice'] / df_subset['livingAreaSqFt']
```

#### Now, let's create a heat map that shows the areas of Austin with the highest price per square foot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate pricePerSqFt
df_subset['pricePerSqFt'] = df_subset['latestPrice'] / df_subset['livingAreaSqFt']

plt.figure(figsize=(10, 6))
plt.scatter(df_subset['longitude'], df_subset['latitude'], c=df_subset['pricePerSqFt'], cmap='plasma')
plt.colorbar(label='Price per SqFt')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Price per SqFt around Austin, TX')
plt.axis('equal')
plt.show()
```

#### As we can see from this heat map, the area with more yellow/orange seems to be a luxury neighbourhood. We can explore whether there is a relationship present between the distance from that neighbourhood, and the price per square foot of a house, to confirm the theory.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Find the maximum price per square foot
max_price_per_sqft = df_subset['pricePerSqFt'].max()
max_point = df_subset[df_subset['pricePerSqFt'] == max_price_per_sqft].iloc[0]

# Visualize the pricePerSqFt geographically
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_subset['longitude'], df_subset['latitude'], c=df_subset['pricePerSqFt'], cmap='plasma', alpha=0.6, edgecolors='w', linewidth=0.5)
plt.colorbar(label='Price per SqFt')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Price per SqFt around Austin, TX')
plt.axis('equal')

# Highlight the highest priced point
plt.scatter(max_point['longitude'], max_point['latitude'], color='red', edgecolors='black', linewidth=2, label='Highest Price per SqFt')
plt.legend()
plt.show()
```

#### Now that we have identified the house with the highest price per square foot, we can create a linear regression model around price per square foot and the distance from the point identified in the graph above.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Function to calculate Euclidean distance
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

# Calculate distance from the highest priced home
df_subset['distanceFromMax'] = df_subset.apply(lambda row: euclidean_distance(max_point['latitude'], max_point['longitude'], row['latitude'], row['longitude']), axis=1)

# Perform regression analysis
X = df_subset[['distanceFromMax']]
y = df_subset['pricePerSqFt']

model = LinearRegression()
model.fit(X, y)
r_squared = model.score(X, y)

print("Intercept:", model.intercept_)
print("Coefficient:",model.coef_[0])
print("R-squared:",r_squared)

plt.figure(figsize=(10, 6))
plt.scatter(df_subset['distanceFromMax'], df_subset['pricePerSqFt'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.plot(df_subset['distanceFromMax'], model.predict(X), color='red', linewidth=2)
plt.xlabel('Distance from Highest Priced Home')
plt.ylabel('Price per SqFt')
plt.title('Regression of Price per SqFt on Distance from Highest Priced Home')
plt.show()
```

#### This plot shows there is a clear relationship between the price of a house and the distance away from that point. As you get further away from the point, the price of the homes go down. The r-squared value here indicates that 38.6% of the variability in price per square foot can be explained by the distance that the house is from the central point.

**We perform regressions for # of price changes to investigate the relationship with price**


```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

#linear regression for price changes
Y_Price = df_subset['latestPrice']
X_PriceChange = df_subset[['numPriceChanges']]

PriceChange_lin_model = LinearRegression()
PriceChange_lin_model.fit(X_PriceChange,Y_Price)

Y_PriceChange_lin_model_pred = PriceChange_lin_model.predict(X_PriceChange)

#model characteristics/statistics
PriceChange_lin_model_b0 = PriceChange_lin_model.intercept_
PriceChange_lin_model_b1 = PriceChange_lin_model.coef_[0]

print("Price Change Linear Regression Intercept: ", PriceChange_lin_model_b0)
print("Price Change Linear Regression Coefficient: ", PriceChange_lin_model_b1)
print("Price Change Linear Regression R-Squared: ", r2_score(Y_Price, Y_PriceChange_lin_model_pred))

print("Price Change Linear Regression MAE: ", mean_absolute_error(Y_Price, Y_PriceChange_lin_model_pred))
print("Price Change Linear Regression MSE: ", mean_squared_error(Y_Price, Y_PriceChange_lin_model_pred))
print("Price Change Linear Regression RMSE: ", mean_squared_error(Y_Price, Y_PriceChange_lin_model_pred, squared=False))
```

More price changes lead to lower prices. Very inaccurate model with large errors.

```python
#polynominal regression for price changes
X_PriceChange_copy = df_subset[['numPriceChanges']].copy()
X_PriceChange_copy['PriceChange_squared'] = X_PriceChange_copy['numPriceChanges']**2

PriceChange_poly_model = LinearRegression()
PriceChange_poly_model.fit(X_PriceChange_copy,Y_Price)

Y_PriceChange_poly_model_pred = PriceChange_poly_model.predict(X_PriceChange_copy)

#model characteristics/statistics
PriceChange_poly_model_b0 = PriceChange_poly_model.intercept_
PriceChange_poly_model_b1 = PriceChange_poly_model.coef_[0]

print("Price Change Polynominal Regression Intercept: ", PriceChange_poly_model_b0)
print("Price Change Polynominal Regression Coefficient: ", PriceChange_poly_model_b1)
print("Price Change Polynominal Regression R-Squared: ", r2_score(Y_Price, Y_PriceChange_poly_model_pred))

print("Price Change Polynominal Regression MAE: ", mean_absolute_error(Y_Price, Y_PriceChange_poly_model_pred))
print("Price Change Polynominal Regression MSE: ", mean_squared_error(Y_Price, Y_PriceChange_poly_model_pred))
print("Price Change Polynominal Regression RMSE: ", mean_squared_error(Y_Price, Y_PriceChange_poly_model_pred, squared=False))
```

More price changes lead to lower prices. More accurate than linear but still inaccurate.

**Predict price based on neighbours through a location-oriented KNN regressor.**

```python
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier

#KNN regression for longitude & latitude
X_Long_Lat = df_subset[['longitude','latitude']]

Long_Lat_knn_reg = KNeighborsRegressor(n_neighbors=3, metric = 'manhattan')
Long_Lat_knn_reg.fit(X_Long_Lat,Y_Price)

Y_Long_Lat_knn_reg_pred = Long_Lat_knn_reg.predict(X_Long_Lat)

#model characteristics/statistics
print("Regression R-Squared: ", r2_score(Y_Price, Y_Long_Lat_knn_reg_pred))

print("KNN Regression MAE:", mean_absolute_error(Y_Price, Y_Long_Lat_knn_reg_pred))
print("KNN Regression MSE:", mean_squared_error(Y_Price, Y_Long_Lat_knn_reg_pred))
print("KNN Regression RMSE:", mean_squared_error(Y_Price, Y_Long_Lat_knn_reg_pred, squared=False))
```

KNN based on location (neighbours and neighbourhoods) is much more accurate with a 70.4% R^2. The error stats are lowest when k=3 and manhattan distance is used (ideal for ordinal data). The model is accurate and is a frequent valuation metric for real estate agents/investors (neighbours are easy comps).

```python
# Check VIF for multi-collinearity
Long_Lat_vif = pd.DataFrame()
Long_Lat_vif["Variable"] = X_Long_Lat.columns

Long_Lat_vif["VIF"] = [variance_inflation_factor(X_Long_Lat.values, i) for i in range(X_Long_Lat.shape[1])]
Long_Lat_vif
```
